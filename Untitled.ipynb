```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get team wins from years 2004 to 2017
# Want to see if having an all star player(s) affect win ratesss
for i in range(2004,2017):
    dfs = pd.read_html('https://www.teamrankings.com/nba/stat/win-pct-all-games?date='+ str(i) +'-06-30', header = 0)
    for df in dfs:
        print(df)
        df.to_csv('team_scores'+ str(i) +'.csv',index = False, encoding ='utf-8')    
```

```python
# Removing and rename columns

for i in range(2004,2017):
    file_path='team_scores'+ str(i)+'.csv'
    data_frame = pd.read_csv(file_path,index_col = False)
    column_name = ['Rank',str(i-2)]
    data_frame= data_frame.drop(column_name,axis=1)
    data_frame = data_frame.rename(columns={str(i):'Win %'})
    data_frame = data_frame.rename(columns={str(2004-1):'Win %','Last 3': 'Last 3 Games Win %', 'Last 1': 'Last Game Win %'})
    data_frame.insert(0,'Year', str(i))
    data_frame.to_csv(file_path, index=False)
```

```python
Indiana = pd.read_csv("Indiana.csv")
Indiana

```

```python
x= Indiana['Year'].tolist()
y= Indiana['Win %'].tolist()

y = ['%.3f' % elem for elem in y]
y = [float(elem) for elem in y]

home_win = Indiana['Home'].tolist()
away_win = Indiana.iloc[:,-1].tolist()

away_win
```

```python
plt.plot(x,y)
plt.xlim(2004,2016)
plt.ylim(0,1)

plt.xlabel('Years')
plt.ylabel('Win Percentage')
plt.title('Indiana\'s Win Percentage from 2004 to 2016' )

plt.show()
```

```python
y_series = pd.Series(y)
y_series.pct_change()

table1 = (y_series.pct_change())
type(table1)

table1 = table1.to_frame()
table1

```

```python
plt.plot(x,home_win, label = 'Home winrate')
plt.plot(x,away_win, label = 'Away winrate')

plt.xlabel('Years')
plt.ylabel('Win Percentage')
plt.title('Indiana\'s Win Percentage from 2004 to 2016' )


plt.legend()

plt.show()
```

```python
# Show the changes of win pct for home wins and away wins

home_win_series = pd.Series(home_win)
print(home_win_series.pct_change())

away_win_series = pd.Series(away_win)
away_win_series.pct_change()


```
